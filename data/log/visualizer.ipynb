```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# load data: 
calib_traj_robot = pd.read_csv('calib_traj_robot.csv', sep=';', index_col=None).to_numpy()
calib_traj_sensor = pd.read_csv('calib_traj_sensor.csv', sep=';', index_col=None).to_numpy()
uncalib_traj_robot = pd.read_csv('trajectory_robot.csv', sep=';', index_col=None).to_numpy()
uncalib_traj_sensor = pd.read_csv('trajectory_sensor.csv', sep=';', index_col=None).to_numpy()
gt = pd.read_csv('gt.csv', sep=';', index_col=None).to_numpy()


# fig 1

plt.plot(gt[:, 0], gt[:, 1], '--', label='ground truth')
plt.plot(calib_traj_sensor[:, 0], calib_traj_sensor[:, 1], label='calib sensor traj', linewidth=2)

plt.legend()

# fig 2

plt.figure(figsize=(10,10)) 
plt.plot(calib_traj_robot[:, 0], calib_traj_robot[:, 1], label='calib robot traj')
plt.plot(calib_traj_sensor[:, 0], calib_traj_sensor[:, 1], '--', label='calib sensor traj')

plt.plot(uncalib_traj_robot[:, 0], uncalib_traj_robot[:, 1], label='uncalib robot traj', color='purple')
plt.plot(uncalib_traj_sensor[:, 0], uncalib_traj_sensor[:, 1], '--', label='uncalib sensor traj')
plt.legend()

# fig 3
chi = pd.read_csv('stats.csv', skiprows=1, header=0)
plt.plot(chi)
```
